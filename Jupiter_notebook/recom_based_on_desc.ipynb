```python
import re
def clean_list(lst):
    return [re.sub(r'[^a-zA-Z0-9\s]', '', word) for word in lst]
new_games_data['description'] = new_games_data['description'].apply(clean_list)
new_games_data
```

```python
def clean_list(lst):
    # Remove non-alphanumeric characters and filter out empty strings
    cleaned_list = [re.sub(r'[^a-zA-Z0-9\s]', '', word).strip() for word in lst]
    # Filter out any empty strings from the list
    return [word for word in cleaned_list if word]

# Apply the cleaning function to each list in the 'desc' column
new_games_data['description'] = new_games_data['description'].apply(clean_list)
```

```python
# new_games_data.to_csv('new_games_data_withArryDesc.csv', index=False)
```

```python
new_games_data['Tags']=new_games_data['description'] + new_games_data['tags']
new_games_data
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['Tags']=new_df['Tags'].apply(stem)
new_df
```

```python
new_df.to_csv('games_preprocessed_with_tags_porterstemmer.csv',index=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vactors=cv.fit_transform(new_df['Tags'])
vactors.shape
```

```python
from scipy.sparse import csr_matrix
sparse_vectors = csr_matrix(vactors)
# print("Shape of sparse matrix:", sparse_vectors.shape)

# # Check memory usage
# data_size = sparse_vectors.data.nbytes
# indices_size = sparse_vectors.indices.nbytes
# indptr_size = sparse_vectors.indptr.nbytes

# total_memory_usage = data_size + indices_size + indptr_size
# print("Memory used by sparse matrix (in mb):", total_memory_usage/(1024*1024))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(sparse_vectors, dense_output=False)
# similarity.shape
memory_usage = similarity.data.nbytes + similarity.indices.nbytes + similarity.indptr.nbytes
print("Memory used by sparse similarity matrix (in bytes):", memory_usage)
print(similarity.shape)
```
