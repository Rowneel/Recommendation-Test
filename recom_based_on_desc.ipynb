```python
new_games_data['tags']=new_games_data['tags'].apply(lambda x:x.split())
new_games_data['tags']
```

```python
new_games_data['description']=new_games_data['description'].apply(lambda x:[i.replace(" ","") for i in x])
new_games_data
```

```python
import re
def clean_list(lst):
    return [re.sub(r'[^a-zA-Z0-9\s]', '', word) for word in lst]
new_games_data['description'] = new_games_data['description'].apply(clean_list)
new_games_data
```

```python
def clean_list(lst):
    # Remove non-alphanumeric characters and filter out empty strings
    cleaned_list = [re.sub(r'[^a-zA-Z0-9\s]', '', word).strip() for word in lst]
    # Filter out any empty strings from the list
    return [word for word in cleaned_list if word]

# Apply the cleaning function to each list in the 'desc' column
new_games_data['description'] = new_games_data['description'].apply(clean_list)
```

```python
# new_games_data.to_csv('new_games_data_withArryDesc.csv', index=False)
```

```python
new_games_data['Tags']=new_games_data['description'] + new_games_data['tags']
new_games_data
```

```python
new_df=new_games_data[['app_id','title','Tags']]
new_df
```

```python
new_df['Tags']=new_df['Tags'].apply(lambda x:" ".join(x))
new_df
```

```python
# list_rows = new_df['Tags'].apply(lambda x: isinstance(x, list))

# Display the rows that contain lists
# print(new_df[list_rows])
new_df['Tags']=new_df['Tags'].apply(lambda x:x.lower())
new_df
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df['Tags']=new_df['Tags'].apply(stem)
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vactors=cv.fit_transform(new_df['Tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vactors)
```

```python
def recommend(game):
    index = new_df[new_df['Name'] == game].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].Name)
```

```python
pickle.dump(similarity,open('similarity.pkl','wb'))
```
